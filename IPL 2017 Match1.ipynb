```python
Extract required columns into a variable 'a'
```

```python
a=data[['batsman','total_runs','ball','match_id']]
```

```python
a
```

```python
#Create New Data Frame NF with required columns and index range
```

```python
NF=pd.DataFrame(index=range(0,248),columns=['Batsman','match-id','Total_Runs','balls','Strike Rate'])
```

```python
Successfully created new data frame and we can empty data frame below.
```

```python
NF
```

```python
#Extract only records of match1
```

```python
b=a.loc[a['match_id'] == 1]
```

```python
b
```

```python
#Copy column 'batsman' from variable b to New Data Frame NF
```

```python
NF['Batsman']=b['batsman']
```

```python
#Copy column 'total_runs' from variable b to New Data Frame NF
```

```python
NF['Total_Runs']=b['total_runs']
```

```python
#Copy column 'ball' from variable b to New Data Frame NF
```

```python
NF['balls']=b['ball']
```

```python
#Copy column 'match_id' from variable b to New Data Frame NF
```

```python
NF['match-id']=b['match_id']
```

```python
NF
```

```python
# Total runs scored, no. of balls played by all batsman in match-id=1
```

```python
NF=NF.groupby(['Batsman','match-id']).sum()
```

```python
NF
```

```python
# Total runs scored, no. of balls, Strike rate played by all batsman in match-id=1
```

```python
NF['Strike Rate']=(NF['Total_Runs']/NF['balls'])*100
```

```python
NF
```

```python
#Rounding Off Strike Rate
```

```python
NF['Strike Rate']=NF['Strike Rate'].apply(lambda x: round(x, 2))
```
